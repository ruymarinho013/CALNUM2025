```python
# M√©todo da Bisse√ß√£o
# Entrada
a = 1 # intervalo [a , b]
b = 3
tolerancia = 0.01 # toler√¢ncia
nloop = 50 # n√∫mero m√°ximo de loop
def f(x):
    return x**2 - 3

import math
from tabulate import tabulate

print("M√©todo da Bisse√ß√£o")

data = []
headers = ["n", "a", "b", "x", "f(a)", "f(b)", "f(x)", "f(a)*f(x)", "err"]

n = 1
fa = f(a)
fb = f(b)
xm2 = (a + b)/2
fxm = f(xm2)
v = fa*fxm
erro = 10
data.append([n, a, b, xm2, fa, fb, fxm, v, erro])

if v < 0:
    b = xm2
if v > 0:
    a = xm2
if v == 0:
    print("o valor da raiz √© %4.4f" %xm2)
else:
  while(erro > tolerancia):
      xm1 = xm2
      n = n + 1.
      xm2 = (a + b)/2
      fxm = f(xm2)
      erro = math.fabs(xm1 - xm2)
      v = fa*fxm
      if v < 0:
          b = xm2
      if v > 0:
          a = xm2
      data.append([n, a, b, xm2, fa, fb, fxm, v, erro])
      if(n == nloop):
          break

print(tabulate(data, headers=headers, floatfmt=".4f"))

print("\nA raiz aproximada √© %4.4f" %xm2)
print('Loop para com no m√°ximo 50 intera√ß√µes')

import matplotlib.pyplot as plt
import numpy as np
xi = np.linspace(-10, 10, 100)
fig = plt.figure()
plt.plot(xi, f(xi), '-')
plt.grid()
```

```python
#M√©todo de Newton
# Entrada
a = float(input("Digite o valor de 'a' (intervalo inicial)")) # intervalo [a , b]
b = float(input("Digite o valor de 'b' (intervalo final)"))
tolerancia = float(input("Digite o valor da Toler√¢ncia:")) # toler√¢ncia
nloop = int(input("Digite o n√∫mero m√°ximo de intera√ß√µes:")) # n√∫mero m√°ximo de loop
#vari√°veis de f(x)
k = float(input("Digite o valor de a na fun√ß√£o:"))
i = float(input("Digite o valor de b na fun√ß√£o:"))
j = float(input("Digite o valor de c na fun√ß√£o:"))
exp = float(input("Digite o valor de expoente na fun√ß√£o:"))

# k*x**exp+i*x+j
def f(x):
    return k*(x**exp)+(i*x)+j

def der1(x):
  #derivada de primeira ordem
  dxd1 = 0.0001
  return (f(x+dxd1)-f(x))/dxd1

def der2(x):
  #derivada de segunda ordem
  dxd2 = 0.0001
  d11 = (f(x+dxd2)-f(x))/dxd2
  d12 = ( f(x + dxd2) - f(x)) / dxd2
  return (d12 - d11) / dxd2

import math
from tabulate import tabulate

print("M√©todo de Newton")

data = []
headers = ["n", "a", "b", "xn", "f(xn)", "erro"]

vfa = 0
vfb = 0
vder2a = 0
vder2b = 0

if(f(a)>= 0):
  vfa = 1
if(f(b) >=0):
  vfb = 1
if(der2(a) >= 0):
  vder2a = 1
if(der2(b) >= 0):
   vder2b = 1
if (vder2a ==vfa):
  xo = a
  c = b
if(vder2b == vfb):
  xo = b
  c = a

#outras vari√°veis
para = 0
xk = 0
h = 0

erro = 10

if(vder2a == vfa):
  data.append([h, xo, c, xo, f(xo), erro])
if (vder2b == vfb):
  data.append([h, c, xo, xo, f(xo), erro])

while(para == 0):
  xk = xo - (f(xo)/der1(xo));
  erro = abs(xk - xo)
  xo = xk
  h = h+1
  if(vder2a == vfa):
    data.append([h, xo, c, xo, f(xo), erro])
  if (vder2b == vfb):
    data.append([h, c, xo, xo, f(xo), erro])

  if(erro < tolerancia):
    if(h>1):
      para = 1

print(tabulate(data, headers=headers, floatfmt=".4f"))

print("\nA raiz aproximada √© %4.4f \n" %xo)
print('Loop para com no m√°ximo 50 intera√ß√µes')

import matplotlib.pyplot as plt
import numpy as np
xi = np.linspace(-10, 10, 100)
fig = plt.figure()
plt.plot(xi, f(xi), '-', label='f(x)')
plt.plot(xi, der1(xi), '--', label="f'(x)")
plt.plot(xo, f(xo), 'ro', label='Root')
plt.plot(xo, der1(xo), 'bo')
plt.annotate(f'Root: {xo:.4f}', (xo, f(xo)),
             textcoords="offset points", xytext=(0,10), ha='center')
plt.grid()
plt.legend()
plt.show()

```

##M√©todo de Newton Simplificado

```python
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate

print("=== M√©todo de Newton-Raphson ===")
a = float(input("Digite o coeficiente a: "))
b = float(input("Digite o coeficiente b: "))
c = float(input("Digite o coeficiente c: "))
expoente = float(input("Digite o expoente: "))
x0 = float(input("Digite o valor inicial (x0): "))
tolerancia = float(input("Digite a toler√¢ncia: "))
max_inter = int(input("Digite o n√∫mero m√°ximo de itera√ß√µes: "))

#Fun√ß√£o e derivada

def f(x):
  return a*(x**expoente)+ b*x + c

def f_derivada(x):
  return a*expoente*(x**(expoente-1)) + b

tabela = []
valores_x = [x0]

#intera√ß√µes do m√©todo
for i in range(max_inter):
  x1 =x0 - f(x0)/f_derivada(x0)
  erro =abs(x1 - x0)
  tabela.append([i+1, x0, f(x0), erro])
  valores_x.append(x1)

  if erro < tolerancia:
    raiz = x1
    break
  x0 =x1
else:
  raiz = x1

#exibe a tabela
headers = ["Itera√ß√£o", "x", "f(x)", "Erro"]
print(tabulate(tabela, headers=headers, floatfmt=".4f"))

print(f"A raiz aproximada √© {raiz:.4f}\n")

#Gr√°fico
xi = np.linspace(min(valores_x)-2, max(valores_x)+2, 200)
yi = f(xi)

plt.figure(figsize = (8,5))
plt.plot(xi, yi, label='f(x)', color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.scatter(raiz, f(raiz), color='red', zorder=5,
           label=f'Raiz aproximada {raiz:.4f}')
plt.title('M√©todo de Newton-Raphson')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
